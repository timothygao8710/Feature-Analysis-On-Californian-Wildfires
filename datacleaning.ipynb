```python
a = pd.read_csv('satellite_fires2.csv')
b = pd.read_csv('satellite_fires.csv')

print(a.head())
print(b.head())
```

```python
locator = geopy.Nominatim(user_agent='john lenon', timeout=10)
a['LocationString'] = a['latitude'].map(str) + ',' + a['longitude'].map(str)
b['LocationString'] = b['latitude'].map(str) + ',' + b['longitude'].map(str)

```

```python
idx = 0
```

```python
with open('dropa.txt', 'a') as file:
    sz = len(a)
    while idx < sz:
        location = locator.reverse(a['LocationString'][idx])
        address = location.address

        if not " California," in address:
            file.write(str(idx) + ' ' + str(address) + '\n')
        
        # if i % 1000 == 0:
        print("done with " + str(idx))
        idx += 1
```

```python
with open('dropb.txt', 'a') as file:
    sz = len(b)
    while idx < sz:
        location = locator.reverse(b['LocationString'][idx])
        address = location.address

        if not " California," in address:
            file.write(str(idx) + ' ' + str(address) + '\n')
        
        # if i % 1000 == 0:
        print("done with " + str(idx))
        idx += 1
```

```python
print(idx)
print(len(a))

tots = []

with open('dropa.txt') as f:
    lines = f.readlines()
    for l in lines:
        dropidx = l.split(' ')[0]

        tots.append(int(dropidx))

print(len(a))
print(len(tots))

print(a.index)
a = a.drop(tots)
a = a.reset_index(drop=True)
print("now df is only " + str(len(a)))
a.head()
a.to_csv('satellite_fires2_finished.csv')
```
