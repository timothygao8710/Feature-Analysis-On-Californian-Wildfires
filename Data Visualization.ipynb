```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('4Gridded_Small_Fires.csv')
print(df.head())
print(df.columns)
```

```python


def bar(feature):
    mp = {}
    for i in df[feature]:
        if(not i in mp) :
            mp[i] = 0
        mp[i] += 1
    fig, ax = plt.subplots(figsize =(20, 14))
    ax.bar(mp.keys(), mp.values())
    plt.title("Bar Graph of " + feature + " Distribution")
    plt.xlabel(feature, fontsize=18)
    plt.ylabel('Frequency', fontsize=16)
    plt.show()

    
def hist(feature):
#     ax.hist(a, bins = [0, 25, 50, 75, 100])
    fig, ax = plt.subplots(figsize =(20, 14))
    ax.hist(df[feature])
    plt.title("Histogram of " + feature + " Distribution")
    plt.xlabel(feature, fontsize=18)
    plt.ylabel('Frequency', fontsize=16)
    plt.show()

    
bar('MO')
hist('DIFF')
bar('T2M')
```
